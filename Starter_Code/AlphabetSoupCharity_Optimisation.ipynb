```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
# Look at the value counts of APPLICATION_TYPE
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Print the value counts
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`


# Count the occurrences of each unique value in the APPLICATION_TYPE column
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Determine the threshold for replacing with "Other"
threshold = 200

# Create a list of application types to be replaced
application_types_to_replace = application_type_counts[application_type_counts <= threshold].index


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
# Get the value counts of the CLASSIFICATION column
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Print the value counts
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# Get the value counts of the CLASSIFICATION column
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Filter the value counts where count is greater than 1
classification_counts_filtered = classification_counts[classification_counts > 1]

# Print the filtered value counts
print(classification_counts_filtered)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classification_cutoff = 2000

# Get the value counts of the CLASSIFICATION column
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts < classification_cutoff].index

# Replace these classifications with "Other" in the dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure replacement was successful
print(application_df['CLASSIFICATION'].value_counts())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

# List of columns to one-hot encode
columns_to_encode = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 
                     'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']

# Apply one-hot encoding to specified columns
application_df = pd.get_dummies(application_df, columns=columns_to_encode, dtype=int)

# Print the DataFrame to verify the changes
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop(columns='IS_SUCCESSFUL').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

* increase the number of neurons in each layer and add a third hidden layer to make the model more complex..


```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

dim =X_train_scaled.shape[1] 
nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(
     units = 128 , activation = 'relu', input_dim = dim))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(
     units = 64 , activation = 'relu'
))
# Third hidden layer
nn_model.add(tf.keras.layers.Dense(
    units= 32, activation='relu'))
# Output layer
nn_model.add(tf.keras.layers.Dense(
     units = 1 , activation = 'sigmoid'
))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']  
)
```

* Adding more epochs to the training regimen.

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs = 200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save("AlphabetSoupCharity_Optimisation.h5")
```
